```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf
```

```python
url = 'https://www.fundamentus.com.br/resultado.php'
agent = {'User-Agent':'Mozzila/5.0'}
resposta = requests.get(url, headers=agent)
soup = BeautifulSoup(resposta.text, 'lxml')
tabela = soup.find_all('table')[0]
df = pd.read_html(str(tabela), decimal=',', thousands='.')[0]
df
```

```python
df.info()
```

```python
# filtro de liquidez maior que 1mm
liquidezMaior = df[df['Liq.2meses'] >= 1000000]
liquidezMaior
```

```python
# filtro de ações sem dividas
patrimonioMaior = liquidezMaior[liquidezMaior['Patrim. Líq'] >= 0]
patrimonioMaior
```

```python
dadosFiltrados = patrimonioMaior[['Papel', 'ROIC', 'EV/EBIT']]
```

```python
dadosFiltrados.head()
```

```python
dadosFiltrados.info()
```

```python
ordemEV = dadosFiltrados.sort_values(by='EV/EBIT')
ordemEV = ordemEV.reset_index(drop=True)
ordemEV['ordemEV'] = ordemEV.index
ordemEV
```

```python
#tratamento valores
ordemEV['ROIC'] = ordemEV['ROIC'].str.replace(',', '.')
ordemEV['ROIC'] = ordemEV['ROIC'].str.rstrip('%').astype('float') / 100
```

```python
ordemEV.dtypes
```

```python
ordemROIC = ordemEV.sort_values(by='ROIC', ascending=False)
ordemROIC = ordemROIC.reset_index(drop=True)
ordemROIC['ordemROIC'] = ordemROIC.index
ordemROIC
```

```python
dadosFinal = ordemROIC
dadosFinal['Score'] = dadosFinal['ordemEV'] + dadosFinal['ordemROIC']
dadosFinal = dadosFinal.sort_values(by='Score')
```

```python
dadosFinal.head()
```

```python
def setorAtivo(tick):
    setor = yf.Ticker(tick)
    return setor.info['sector']
```

```python
tickers = dadosFinal['Papel'] + '.SA'
tickers
```

```python
dicionarioSetores = {}
for t in tickers:
    try:
        dicionarioSetores[t] = setorAtivo(t)
        print('COLETA DE DADOS')
    except:
        print(f'ATIVO SEM COLETA : {t}')
        dicionarioSetores[t] = 'NAN'

dicionarioSetorAtivo = pd.DataFrame.from_dict(dicionarioSetores, orient='index').reset_index()
dicionarioSetorAtivo.columns = ['Papel', 'Setor']

dicionarioSetorAtivo
```
