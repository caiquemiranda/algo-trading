```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dolar = pd.read_excel('dolar_4x.xlsx')
dolar.set_index('Data', inplace = True)
dolar['balance'] = (dolar['Fechamento'] - dolar['Abertura']) / (dolar['Máxima'] - dolar['Mínima'])
dolar['amplitude(pts)'] = (dolar['Fechamento'] - dolar['Abertura'])
dolar['amplitude(total)'] = (dolar['Máxima'] - dolar['Mínima'])
act=[]
for i in dolar['balance']:
    if i >= 0.7:
        act.append('venda')
    elif i <= -0.7:
        act.append('compra')
    else:
        act.append('hold')
dolar['act'] = act
```

```python
dolar.drop('ULT', axis = 1, inplace = True)
dolar.drop('Mínima', axis = 1, inplace = True)
dolar.drop('Máxima', axis = 1, inplace = True)
dolar.drop('Abertura', axis = 1, inplace = True)
```
